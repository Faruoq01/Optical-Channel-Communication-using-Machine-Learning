```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
sb.set()
```

# Import Data

```python
data = pd.read_csv('Experimental_Data.csv')
data
```

```python
plt.scatter(data['Visibility'], data['value q'], c='b')
plt.xlabel('Visibility')
plt.ylabel('q')
```

# Set Inputs and Outputs

```python
inputs = np.column_stack((data['Visibility'],))
targets = np.column_stack((data['value q'],))
```

# Initialize Variables

```python
Learning_Rate = 0.01
Weight = np.random.uniform(0, 1, size=1)
Bias = np.random.uniform(0, 1, size=1)
obs = 50
```

# Optimization Algorithm

```python
for i in range(500):
    
    ## make a random prediction from the random guess value

    outputs = Weight*np.log(inputs) + Bias
    delta = outputs - targets

    ## calculate the loss function (objective function)
    Loss = np.sum(delta**2)/obs
    print(Loss)

    delta_scaled = 2*delta/obs
    
    ## updates the weight and Bias
    Weight = Weight - Learning_Rate*np.dot(np.log(inputs).T, delta_scaled)
    Bias = Bias - Learning_Rate*np.sum(delta_scaled)
```

# Trained Model

```python
outputs = Weight*np.log(inputs) + Bias
plt.scatter(inputs, targets, c='b')
plt.plot(inputs, outputs, c='r')
plt.xlabel('Visibility')
plt.ylabel('q')
```

# Compare All Models

# Kruse Model

```python
lambda1 = 5.5e-7
lambda2 = 1.55e-6
consta = 3.19

kruse_visibility = data_atten['Visibility']
kruse_q = 0.585*kruse_visibility**(1/3)
#print(kruse_q)
atten1 = consta*(lambda1/lambda2)**kruse_q
kruse_attenuation = atten1/kruse_visibility
#print(kruse_attenuation)
##plt.scatter(data_atten['Visibility'], data_atten['Attenuation'], c='r')
plt.plot(data_atten['Visibility'], kruse_attenuation, c='b')
plt.xlabel('Visibility')
plt.ylabel('Attenuation')
print(kruse_attenuation)
```

# Kim Model

```python
lambda1 = 5.5e-7
lambda2 = 1.55e-6
consta = 3.19

kim_visibility = data_atten['Visibility']
kim_q = 0

atten1 = consta*(lambda1/lambda2)**kim_q
kim_attenuation = atten1/kim_visibility
print(kim_visibility)
##plt.scatter(data_atten['Visibility'], data_atten['Attenuation'], c='r')
plt.plot(kim_visibility, kim_attenuation, c='y')
plt.xlabel('Visibility')
plt.ylabel('Attenuation')
```
