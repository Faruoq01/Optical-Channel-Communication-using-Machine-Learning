# Rain Data (Dust)


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
sb.set()
```

```python
Visibility_range = np.linspace(2.8, 4, 60)
Noise = np.random.uniform(0, 0.2, 60)
attenuation = 2.8/Visibility_range + Noise


plt.scatter(Visibility_range, attenuation)
```

# Machine Learning Data

```python
inputs = np.column_stack((Visibility_range,))
Targets = np.column_stack((attenuation,))
```

# Initialization

```python
Learning_rate = 0.1
Weight = np.random.uniform(0, 1, size=1)
Bias = np.random.uniform(0, 1, size=1)
obs = 60
```

# Optimization Algorithm

```python
for i in range(100):
    outputs = Weight/inputs
    delta = outputs - Targets
    Loss = np.sum(delta**2)/obs
    print(Loss)
    
    delta_scaled = 2*(delta)/obs
    Weight = Weight - Learning_rate*np.sum(inputs*delta_scaled)
```

# Testing

```python
model = Weight/inputs

plt.scatter(Visibility_range, attenuation, c='r')
plt.plot(inputs, model)
plt.xlabel('Visibility')
plt.ylabel('Attenuation')
```

# Comparison with Kaushal Model

```python
plt.scatter(Visibility_range, attenuation, c='r')
plt.plot(Visibility_range, 2.8/Visibility_range, c='green')
plt.plot(inputs, model)
plt.xlabel('Visibility')
plt.ylabel('Attenuation')
```
