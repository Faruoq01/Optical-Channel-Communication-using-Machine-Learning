# Randomize the data

```python
## Generate random data
lamda_1 = 5.5e-7
lamda_2 = 1.55e-6
x = 3.19
Visibility_range = np.linspace(1, 500, 60)
y = 70.5*np.exp(-Visibility_range/1000)
k = np.log(x) - np.log(y)
a = 1/(np.log(lamda_1/lamda_2))
b = a*k

##Randomize the generated data
Noise = np.random.uniform(-0.5, 0.5, 60)

q = -a*np.log(Visibility_range/1000) + b

print(q)

plt.scatter(Visibility_range/1000, q + Noise, c = 'r' )
```

```python
data = {'value q': q+Noise, 'Visibility':Visibility_range/1000}
df = pd.DataFrame(data, columns = ['value q', 'Visibility'])
print(df)
```
